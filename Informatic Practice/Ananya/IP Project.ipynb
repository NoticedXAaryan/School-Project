# <Center> Analyzing the Trends of Fine Particulate Matter (PM2.5) Concentrations in Urban and Rural India (2010-2019)

## Source for Data

This data was collected form WHO [ https://www.who.int/data/gho/data/themes/air-pollution/who-air-quality-database ] 

## About Data

The data shows the concentrations of fine particulate matter, specifically PM2.5, in India from two thousand ten to two thousand nineteen. PM2.5 refers to airborne particles with a diameter of less than two point five micrometers. These particles are significant because they can penetrate deep into the lungs and even enter the bloodstream, causing various health problems.

#### what the data means:

1. **Health Impact**: High levels of PM2.5 are linked to respiratory and cardiovascular diseases. The World Health Organization considers PM2.5 a major risk factor for premature death. Therefore, the values in your data reflect the potential health risks for the population.

2. **Categories**: The data is segmented into different categories:
   - **Total**: The overall concentration across all areas.
   - **Urban**: Concentrations specifically in urban areas, where pollution is typically higher due to traffic, industry, and construction.
   - **Rural**: Concentrations in rural areas, generally lower due to less industrial activity.

3. **Trends Over Time**: The data shows trends over the years. For instance, while PM2.5 concentrations in some years (like two thousand thirteen) are relatively high, there is a noted decrease in recent years (like two thousand nineteen). This decline could indicate improvements in air quality management, regulations, or public awareness about pollution control.

4. **Urban vs. Rural Differences**: Urban areas consistently report higher PM2.5 levels than rural areas, highlighting the environmental challenges cities face. This difference is crucial for policymakers, as it helps identify where to focus air quality improvement efforts.

5. **Source Reference**: The source you provided likely contains additional context, such as health recommendations, the impact of specific policies, or comparisons with other countries. These aspects can provide a broader understanding of air quality management in India and how it compares globally.

```python
# Importing Reqiered library
import pandas as pd
from matplotlib import pyplot as plt
```

### Create a DataFrame and a CSV

```python
# Defining the data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Urban': [68.4, 65.7, 63.1, 61.9, 60.8, 58.7, 57.3, 56.4, 55.0, 53.8],
    'Rural': [45.2, 44.1, 43.0, 42.5, 40.8, 39.7, 39.3, 38.5, 37.1, 35.9],
    'Total': [55.3, 53.9, 51.4, 50.4, 48.4, 47.5, 45.5, 44.5, 43.1, 42.6]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Exporting CSV 
df.to_csv(".CSV", index = False)   # will be stored in the same folder.
```

*************************************************************************************************************************************************

## Analyzing and Visualizing the Data

```python
# Single line graph for Urban and Rural PM2.5 Concentration
plt.plot(df['Year'], df['Urban'], marker='o', label='Urban', color='blue')
plt.plot(df['Year'], df['Rural'], marker='o', label='Rural', color='green')

plt.title('PM2.5 Concentration in Urban and Rural Areas (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Multiple line graphs for Urban, Rural, and Total PM2.5 Concentration
plt.plot(df['Year'], df['Urban'], marker='o', label='Urban', color='blue')
plt.plot(df['Year'], df['Rural'], marker='o', label='Rural', color='green')
plt.plot(df['Year'], df['Total'], marker='o', label='Total', color='orange')

plt.title('PM2.5 Concentration in Urban, Rural, and Total Areas (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Multiple bar graphs for Total PM2.5 Concentration by Year
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Total'], color='orange')

plt.title('Total PM2.5 Concentration by Year (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a bar graph for Urban PM2.5
plt.bar(df['Year'], df['Urban'], color='b')
plt.title('PM2.5 Concentration in Urban Areas (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')  # BTW (g/m) was taken from the WHO website
plt.show()
```

```python
# Plot a bar graph for Rural PM2.5
plt.bar(df['Year'], df['Rural'], color='g')
plt.title('PM2.5 Concentration in Urban Areas (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')  # BTW (g/m) was taken from the WHO website
plt.show()
```

```python
# Set positions for bars
x = range(len(df['Year']))

# Plot multiple bar graphs for Urban and Rural
plt.bar(x, df['Urban'], width=0.4, label='Urban', color='blue', align='center')
plt.bar([p + 0.4 for p in x], df['Rural'], width=0.4, label='Rural', color='green', align='center')
plt.title('PM2.5 Concentration in Urban vs Rural Areas (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')  # BTW (g/m) was taken from the WHO website
plt.xticks([p + 0.2 for p in x], df['Year'])
plt.legend()
plt.show()
```

```python
# Set positions for bars
x = range(len(df['Year']))

# Plot multiple bar graphs for Urban, Rural, and Total
plt.bar(x, df['Urban'], width=0.25, label='Urban', color='blue', align='center')
plt.bar([p + 0.25 for p in x], df['Rural'], width=0.25, label='Rural', color='green', align='center')
plt.bar([p + 0.5 for p in x], df['Total'], width=0.25, label='Total', color='orange', align='center')

plt.title('Comparison of PM2.5 Concentration: Urban, Rural, and Total (2010-2019)')
plt.xlabel('Year')
plt.ylabel('PM2.5 Concentration (g/m)')
plt.xticks([p + 0.25 for p in x], df['Year'])
plt.legend()
plt.show()
```

```python
# Calculate total PM2.5 for Urban and Rural
urban_total = df['Urban'].sum()
rural_total = df['Rural'].sum()

# Pie chart
labels = ['Urban', 'Rural']
sizes = [urban_total, rural_total]
colors = ['blue', 'green']
explode = (0.1, 0)  # explode the urban slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Total PM2.5 Concentration Contribution (2010-2019)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
