## Analyzing and Visualizing the Data

### Step 1: Create a DataFrame

```python
import pandas as pd

# Real-world COVID-19 data for 2020
data = {
    "Month": ["January", "February", "March", "April", "May", "June","July", "August", "September", "October", "November", "December"],
    
    "Total Cases": [10000, 100000, 500000, 3000000, 6000000, 10000000,16000000, 25000000, 30000000, 36000000, 45000000, 80000000],

    "Total Deaths": [200, 3000, 22000, 200000, 360000, 500000,650000, 800000, 1000000, 1200000, 1400000, 1800000],
    
    "Total Recovered": [1000, 30000, 100000, 800000, 1500000, 5000000,7500000, 12000000, 15000000, 20000000, 25000000, 40000000]
}

# Create the DataFrame
covid = pd.DataFrame(data)

# Exporting CSV as "Covid-19_2020.CSV"
covid.to_csv("Covid-19_2020.CSV", index = False)   # will be stored in the same folder.

# Printing DataFrame
print( covid )
```

## Step 2: Plotting the Data

```python
# Importing the pyplot from matpotlib
import matplotlib.pyplot as plt
```

```python
# Line plot for Total Cases
plt.plot(covid["Month"], covid["Total Cases"], label='Total Cases', marker='o')
plt.title("Total COVID-19 cases in 2020")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Line plot for Total Cases, Deaths, and Recoveries
plt.plot(covid["Month"], covid["Total Deaths"], label='Total Deaths', marker='o')
plt.title("COVID-19 Deaths in 2020")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Line plot for Total Cases, Deaths, and Recoveries
plt.plot(covid["Month"], covid["Total Recovered"], label='Total Recovered', marker='o')
plt.title("COVID-19 Recoveries in 2020")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
