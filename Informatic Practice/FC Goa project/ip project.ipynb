# Creating Data Frame

```python
import pandas as pd

data = {'Season': ['2019-2020', '2020-2021', '2021-2022', '2022-2023', '2023-2024'],
        'Matches_Played': [20, 21, 22, 20, 22],
        'Wins': [12, 10, 9, 8, 14],
        'Losses': [5, 6, 8, 9, 6],
        'Draws': [3, 5, 5, 3, 2]}

df = pd.DataFrame(data)

print(df)
```

# Analysis

```python
import matplotlib.pyplot as plt
```

```python
# Line chart for wins
plt.plot(df['Season'], df['Wins'], marker='o', label='Wins', color='r')
plt.title('FC Goa Wins Over Seasons')
plt.xlabel('Season')
plt.ylabel('Wins')
plt.show()
```

```python
# Line chart for losses
plt.plot(df['Season'], df['Losses'], marker='o', label='Losses', color='g')
plt.title('FC Goa Losses Over Seasons')
plt.xlabel('Season')
plt.ylabel('Losses')
plt.show()
```

```python
# Line chart for draws
plt.plot(df['Season'], df['Draws'], marker='o', label='Draws', color='b')
plt.title('FC Goa Draws Over Seasons')
plt.xlabel('Season')
plt.ylabel('Draws')
plt.show()
```

```python
# Multiple line chart (Wins, Losses, Draws)
plt.plot(df['Season'], df['Wins'], marker='o', label='Wins', color='r')
plt.plot(df['Season'], df['Losses'], marker='o', label='Losses', color='g')
plt.plot(df['Season'], df['Draws'], marker='o', label='Draws', color='b')
plt.title('FC Goa Wins, Losses, and Draws')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# bar chart
df.plot(kind='bar', x='Season', y=['Wins', 'Losses', 'Draws'])
plt.title('FC Goa Wins, Losses, and Draws Comparison')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for Wins
plt.hist(df['Wins'], bins=5, color='g')
plt.title('Distribution of Wins Over Seasons')
plt.xlabel('Wins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart
labels = ['Wins', 'Losses', 'Draws']
sizes = [df.loc[df['Season'] == '2023-2024', 'Wins'].values[0],
         df.loc[df['Season'] == '2023-2024', 'Losses'].values[0],
         df.loc[df['Season'] == '2023-2024', 'Draws'].values[0]]
colors = ['r', 'g', 'b']

plt.pie(sizes, labels=labels, colors=colors)
plt.title('Wins, Losses, and Draws for FC Goa (2023-2024)')
plt.show()
```
