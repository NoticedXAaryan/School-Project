## Filtering the Data

In this step, I filtered the collected datasets to extract the relevant information for my analysis:

1. **Filtering the Consumer Sentiment Index (CSI) Data:**
   - From the *CSI.csv* file, I extracted the CSI values for the years two thousand to twenty twenty. This ensures we only work with the data relevant to our project.

```python
# Programing part of extraction

# Load the data
data = pd.read_csv('CSI.csv') 

# Filter the data for the years 2000 to 2020
filtered_CSI = data[(data['YYYY'] >= 2000) & (data['YYYY'] <= 2020)]

# Save the filtered data to a new CSV file
filtered_CSI.to_csv('Filtered_CSI.csv', index=False) 

# printing the Top and Bottom most 5 data 
print ( filtered_CSI.head() )
print ( filtered_CSI.tail() )
```

The filtered datasets will be used for further analysis to understand the correlation between GDP and the Consumer Sentiment Index.

*************************************************************************************************************************************************

## Comparing the Data

```python
# Merging the datasets

# Load the GDP data
gdp_data = pd.read_csv('GDP.csv')

# Load the CSI data
csi_data = pd.read_csv('Filtered_CSI.csv')

# Display the initial data for reference (optional)
print("GDP Data:")
print(gdp_data.head())
print("\nCSI Data:")
print(csi_data.head())

# Merge the datasets on the year
merged_data = pd.merge(gdp_data, csi_data, left_on='Year', right_on='YYYY')

# Remove the 'Month' and 'YYYY' columns from the merged DataFrame
merged_data.drop(columns=['Month', 'YYYY'], inplace=True)

# Display the merged data
print("\nMerged Data:")
print(merged_data)

# Save the merged data to a new CSV file
merged_data.to_csv('Merged_GDP_CSI.csv', index=False)
```

```python
# Assuming you have your merged DataFrame called 'merged_data'

# Calculate growth rate of GDP
merged_data['Growth_Rate_GDP'] = merged_data['GDP (Billion $)'].pct_change() * 100  # Convert to percentage

# Calculate year-over-year changes in CSI
merged_data['Change_CSI'] = merged_data['ICS_ALL'].diff()

# Drop the first row with NaN values from diff
analysis_data = merged_data.dropna()

# Display the first few rows of the analysis data
print(analysis_data[['Year', 'Growth_Rate_GDP', 'Change_CSI']])
```

```python
# Plotting
plt.figure(figsize=(12, 6))

# Plot growth rate of GDP
plt.plot(analysis_data['Year'], analysis_data['Growth_Rate_GDP'], marker='o', label='Growth Rate of GDP (%)', color='blue')

# Plot change in CSI
plt.plot(analysis_data['Year'], analysis_data['Change_CSI'], marker='o', label='Change in CSI', color='orange')

# Adding titles and labels
plt.title('Growth Rate of GDP vs. Change in Consumer Sentiment Index (CSI)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Adding a horizontal line at zero for reference
plt.legend()
plt.grid()
plt.show()
```

```python
# Improved Plotting with Twin Axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting Growth Rate of GDP on primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Growth Rate of GDP (%)', color='blue')
ax1.plot(analysis_data['Year'], analysis_data['Growth_Rate_GDP'], marker='o', color='blue', label='Growth Rate of GDP (%)')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for Change in CSI
ax2 = ax1.twinx()
ax2.set_ylabel('Change in CSI', color='orange')
ax2.plot(analysis_data['Year'], analysis_data['Change_CSI'], marker='o', color='orange', label='Change in CSI')
ax2.tick_params(axis='y', labelcolor='orange')

# Adding titles, grid, and legends
plt.title('Growth Rate of GDP vs. Change in Consumer Sentiment Index (CSI)')
fig.tight_layout()  # Adjust layout to prevent overlap
ax1.grid()
plt.show()
```
