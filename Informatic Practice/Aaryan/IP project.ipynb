# <center>Student Spending Analysis

This project analyzes student spending habits by allowing them to input there spending in pre-defined categories (like transport, food, etc.). This project aims to collect data from multiple students and then compared and visualized using charts and graphs for deeper insights.

## Index:
1. Collecting Data
2. Data Comparison
3. Visualizing Spending

## Step 1: Collecting Data

In this step, we will collect spending data from multiple students. Each student can input their own spending in difined categories (like transport, food etc.)

```python
# Importing necessary libraries
import pandas as pd
from matplotlib import pyplot as plt

# Ascertaining no. of data to collect.
try:
    students = int(input("Enter the number of students: "))
except ValueError:
    print("Invalid input. Please Re-run and enter a valid integer.")
    exit()

# Define variable to hold the student data
student_data = {}

# Collect data for each student
for i in range(students):
    try: 
        name = input("Enter student name: ")
        food = float(input(f"Enter {name}'s spending on Food: "))
        transport = float(input(f"Enter {name}'s spending on Transport: "))
        entertainment = float(input(f"Enter {name}'s spending on Entertainment: "))
        clothes = float(input(f"Enter {name}'s spending on Clothes: "))
        miscellaneous = float(input(f"Enter {name}'s spending on Miscellaneous: "))
        
        # Adding to dictionary
        student_data[name] = {
            'Food': food,
            'Transport': transport,
            'Entertainment': entertainment,
            'Clothes': clothes,
            'Miscellaneous': miscellaneous
        }
    except ValueError:
        print(f"Invalid input for student {i+1}. Please Re-run and enter valid numbers.")
        exit()

# Converting to DataFrame
spending_df = pd.DataFrame(student_data).T  # Transpose to match student names as rows

print("\nStudent Spending Data:\n", spending_df)
```

## Step 2: Data Comparison

In this step, Creating another DataFrame to summarize student data into features like total, average, and minimum spending, alongside the most spent field

```python
# Data Comparison
summary_df = pd.DataFrame(index=spending_df.index)

# Adding calculated fields
summary_df['Total Spent'] = spending_df.sum(axis=1)
summary_df['Average Spent'] = spending_df.mean(axis=1)
summary_df['Minimum Spent'] = spending_df.min(axis=1)
summary_df['Most Spent Field'] = spending_df.idxmax(axis=1)

# Displaying the summary DataFrame
print("\nSummary Data:\n", summary_df)
```

## Step 3: Visualizing Spending

In this step, we will visualize the spending and summarised data collected. Different types of charts will be used to highlight spending trends and comparisons across categories and students. This visualization will to help in better understanding the spending habits.

```python
# Step 3: Creating pie charts and bar graphs for each student
for student in spending_df.index:
    student_data = spending_df.loc[student]
    
    # Creating pie chart for spending breakdown
    student_data.plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        title=f"{student}'s Spending Breakdown",
        figsize=(6, 6))
    plt.ylabel('')  # Remove default y-label
    plt.show()

    # Creating bar graph for spending categories
    student_data.plot.bar(
        title=f"{student}'s Spending by Category", ylabel="Amount Spent")
    plt.show()
```
