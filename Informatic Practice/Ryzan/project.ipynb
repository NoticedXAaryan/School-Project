# Topic - Messi's Goal, Assists and Total contribution per season.

# Creating Data Frame

```python
import pandas as pd

data = {
    'Season': ['2015/16', '2016/17', '2017/18', '2018/19', '2019/20', '2020/21', '2021/22', '2022/23', '2023 MLS'],
    'Goals': [41, 54, 45, 51, 31, 38, 11, 16, 20],
    'Assists': [23, 17, 18, 22, 25, 12, 15, 16, 10],
    'Total Contributions': [64, 71, 63, 73, 56, 50, 26, 32, 30]
}


df = pd.DataFrame(data)

print(df)
```

# Analysis

```python
import matplotlib.pyplot as plt
```

```python
# Line chart
plt.plot(df['Season'], df['Goals'], marker='o', color='b')
plt.title("Lionel Messi's Goals per Season")
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
# Line chart
plt.plot(df['Season'], df['Assists'], marker='s', color='g')
plt.title("Lionel Messi's Assists per Season")
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
# Line chart
plt.plot(df['Season'], df['Total Contributions'], marker='s', color='r')
plt.title("Lionel Messi's Total Contributions per Season")
plt.xlabel('Season')
plt.ylabel('Total Contributions')
plt.show()
```

```python
# Multiple line chart
plt.plot(df['Season'], df['Goals'], marker='o', label='Goals', color='b')
plt.plot(df['Season'], df['Assists'], marker='o', label='Assists', color='g')
plt.plot(df['Season'], df['Total Contributions'], marker='o', label='Total Contributions', color='r')
plt.title("Lionel Messi's Total Contributions, Goals and Assists per Season")
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Bar chart
df.plot(kind='bar', x='Season', y=['Goals', 'Assists', 'Total Contributions'])
plt.title("Lionel Messi's Goals, Assists and Total Contributions Comparison")
plt.ylabel('Count')
plt.show()
```

```python
# Create a pie chart
x = len(df)
for index in range(x):
    row = df.iloc[index] 
    sizes = [row['Goals'], row['Assists']]
    plt.pie(sizes, labels=['Goals', 'Assists'], colors=['g', 'b'])
    plt.title(f'Goals and Assists for FC Goa ({row["Season"]})')
    plt.show()
```
